```python
from pathlib import Path
import pandas as pd

BASE_DIR = Path("/content/drive/MyDrive/TFM")
DATA_DIR = BASE_DIR / "data"
RUTA_ORIGINAL = DATA_DIR / "TTT.xlsx"
RUTA_PROCESSED = DATA_DIR / "processed"
RUTA_CSV_PERFIL = RUTA_PROCESSED / "perfil_columnas.csv"

print("BASE_DIR:", BASE_DIR)
print("DATA_DIR:", DATA_DIR)
print("RUTA_ORIGINAL:", RUTA_ORIGINAL)
print("RUTA_CSV_PERFIL:", RUTA_CSV_PERFIL)
```

```python
df = pd.read_excel(RUTA_ORIGINAL)
df.shape, df.head()
```

```python
# Filtrar filas resumen por centro
df_totales_centros = df[df["TEMPORALIDAD"] == "Total"].copy()

# Comprobar resultados
print(f"Número de centros únicos: {df_totales_centros['CD_CENTRO'].nunique()}")
display(df_totales_centros.head())
```

```python
# Renombrar la columna 'Unnamed: 0' a 'MesAno'
df_totales_centros.rename(columns={"Unnamed: 0": "MesAno"}, inplace=True)

print(df_totales_centros.columns)
```

```python
# La convertimos en fecha
df_totales_centros["MesAno"] = pd.to_datetime(df_totales_centros["MesAno"])
```

```python
# Guardar en carpeta processed
RUTA_TOTALES = DATA_DIR / "processed" / "TTT_totales_por_centro.xlsx"
df_totales_centros.to_excel(RUTA_TOTALES, index=False)
print(f"✅ Archivo con totales por centro guardado en: {RUTA_TOTALES}")

```
